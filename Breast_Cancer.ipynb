# Navie Bayes Algorithm

```python
data = pd.read_csv("spam.tsv",sep='\t',names=['Class','Message'])
data.head(2)
```

```python
data['Length'] = data['Message'].apply(len)
```

```python
data.head(2)
```

```python
data.head(2)
```

```python
data.loc[(data['Class']=='ham') & (data['Length']>700)]
```

```python
data.shape
```

```python
data.iloc[:, 0:-1]
```

```python
# encoding

data.loc[data['Class']=='ham', 'Class'] = 1
```

```python
data.loc[data['Class']=='spam', 'Class'] = 0
```

```python
data.head(2)
```

```python
import string

string.punctuation
```

```python
def rem_punch(Msg):
  Msg = "".join([text for text in Msg if text not in string.punctuation])
  return Msg

data['Clean_Msg'] = data['Message'].apply(lambda x: rem_punch(x))
data.head(2)
```

# Vectorizer

To convert words to numerical

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV =  CountVectorizer(stop_words='english')
```

```python
X = data['Clean_Msg'].values
y = data['Class'].values
```

```python
y.dtype
```

```python
y = y.astype('int')
```

```python
y.dtype
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
X_train_CV = CV.fit_transform(X_train)
```

```python
X_train_CV
```

```python
X_test_CV = CV.transform(X_test)
```

```python
X_test_CV
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train_CV, y_train)
```

```python
y_pred = model.predict(X_test_CV)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

# TFIFD - Term Frequency Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
TF = TfidfVectorizer()
X = TF.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=6)
```

```python
from sklearn.naive_bayes import BernoulliNB

## model object creation
nb=BernoulliNB(alpha=0.01)

## fitting the model
nb.fit(X_train,y_train)

## getting the prediction
y_hat=nb.predict(X_test)
```

```python
y_hat
```

```python
print(classification_report(y_test,y_hat))
```
